# Maven Halloween Challenge

**Goal**: Use data to find the 3 types of Halloween candy that will make you the most popular house on the block.

**Details**: Using online votes ranking 85 types of candy, your task is to find the 3 treats you'll give out on Halloween to guarantee that trick-or-treaters of **_all tastes_** find something they'll love and present the data to back up your decision.

## My approach
1. **Python**: View the data
2. **Python**: Apply PCA and jitter to prepare the data for a scatter plot
3. **Tableau**: Explore the scatter plot and make recommendations

## 1. View the data

```python
# read in the candy data flat file
import pandas as pd

data = pd.read_csv('candy-data.csv')
data.head()
```

```python
# look only at the 1/0 columns
subset = data.iloc[:, 1:-3]
subset.head()
```

## 2. Apply Principal Component Analysis (PCA)

```python
# import pca to reduce the 9 columns into 2 columns
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(subset)
```

```python
# the 2 columns account for ~60% of the variation in the data
pca.explained_variance_ratio_
```

```python
# view the 2 columns
candy_2d = pd.DataFrame(pca.transform(subset))
candy_2d.head()
```

```python
# plot the 2 columns
candy_2d.plot(kind='scatter', x=0, y=1);
```

```python
# view the original data again - note that one dime and one quarter are identical
data.head()
```

```python
# add a bit of noise to the data and replot the data
import numpy as np

data['x_jitter'] = candy_2d[0] + np.random.randn(85)*.1
data['y_jitter'] = candy_2d[1] + np.random.randn(85)*.1
data.plot(kind='scatter', x='x_jitter', y='y_jitter');
```

```python
# view the updated dataframe with the jitter columns
data.head()
```

```python
# export the data with the 2 new columns
data.to_csv('candy-data-pca.csv')
```
